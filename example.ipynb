## PDF table extraction

This notebook loads **`sample-tables.pdf`** from the **current working directory** (run with the repo root as cwd, or adjust `pdf_path` below).

**Requirements:** Python 3.13+, [`img2table`](https://pypi.org/project/img2table/), **Tesseract** on `PATH` with **`eng`** trained data (see `lang=` on `TesseractOCR`). Install notebook deps with `uv sync --group notebook` (see `pyproject.toml`).

```python
from img2table.document import PDF
from img2table.ocr import TesseractOCR
from IPython.display import display_html
```

```python
from pathlib import Path

pdf_path = Path("sample-tables.pdf")
if not pdf_path.is_file():
    raise FileNotFoundError(
        f"Missing {pdf_path.resolve()}. Add the file or set the kernel cwd to the directory that contains it."
    )

doc = PDF(str(pdf_path))

ocr = TesseractOCR(n_threads=1, lang="eng")

# Table extraction
extracted_tables = doc.extract_tables(
    ocr=ocr,
    implicit_rows=False,
    implicit_columns=False,
    borderless_tables=False,
    min_confidence=50,
)
```

```python
extracted_tables
```

### Render extracted tables

The next cell prints each `ExtractedTable` as HTML via `table.html_repr(...)`.

```python
for page, tables in extracted_tables.items():
    for idx, table in enumerate(tables):
        display_html(table.html_repr(title=f"Page {page + 1} - Extracted table n°{idx + 1}"), raw=True)
```
